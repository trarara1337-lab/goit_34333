```python
# Завдання 1. Підрахунок загальної кількості респондентів

# Використовуємо метод nunique() для колонки ResponseId, 
# щоб переконатися, що ми рахуємо саме унікальних респондентів.
total_respondents = df['ResponseId'].nunique()

print(f"Загальна кількість респондентів в опитуванні: {total_respondents}")
```

```python
# Завдання 2. Аналіз повноти відповідей респондентів

# 1. Отримуємо список унікальних імен питань із файлу схеми (df1)
schema_questions = set(df1['qname'].unique())

# 2. Отримуємо список усіх колонок, що є в основному датасеті (df)
data_columns = set(df.columns)

# 3. Використовуємо інтерсекцію (перетин), щоб знайти тільки ті колонки, 
# які є і в схемі, і в самому датасеті
columns_to_check = list(schema_questions.intersection(data_columns))

print(f"Кількість питань для перевірки (після інтерсекції): {len(columns_to_check)}")

# 4. Фільтруємо основний датасет, залишаючи тільки ці колонки, 
# та видаляємо всі рядки, де є хоча б один пропуск (NaN)
respondents_with_all_answers = df[columns_to_check].dropna()

# 5. Рахуємо кількість таких "ідеальних" респондентів
final_count = len(respondents_with_all_answers)

print(f"Кількість респондентів, які відповіли на ВСІ запитання: {final_count}")
```

```python
# Завдання 3. Статистичний аналіз досвіду респондентів (WorkExp)

# 1. Розраховуємо основні показники
mean_exp = df['WorkExp'].mean()
median_exp = df['WorkExp'].median()
# Метод mode() повертає Series (бо мод може бути кілька), тому беремо перший елемент [0]
mode_exp = df['WorkExp'].mode()[0]

# 2. Створюємо зручну таблицю (Series) для представлення результатів
work_exp_stats = pd.Series({
    'Середнє значення (Mean)': mean_exp,
    'Медіана (Median)': median_exp,
    'Мода (Most frequent)': mode_exp
})

# 3. Виводимо результат
print("Міри центральної тенденції для досвіду роботи (WorkExp):")
print(work_exp_stats)

# Додатково: подивимось, скільки людей взагалі вказали свій досвід
print(f"\nКількість респондентів, що вказали досвід: {df['WorkExp'].notnull().sum()}")
```

```python
# Завдання 4. Аналіз віддаленої роботи

# 1. Подивимось на всі можливі варіанти у колонці RemoteWork
print("Варіанти форматів роботи:")
print(df['RemoteWork'].unique())

# 2. Фільтруємо респондентів, які обрали саме 'Remote'
remote_workers_count = df[df['RemoteWork'] == 'Remote'].shape[0]

# 3. Виводимо фінальний результат
print(f"\nКількість респондентів, які працюють ВІДДАЛЕНО: {remote_workers_count}")
```

```python
# Завдання 5. Визначення популярності Python (Виправлено)

# 1. Загальна кількість респондентів 
total_respondents = len(df)

# 2. Рахуємо кількість тих, хто вказав Python
python_users_count = df['LanguageHaveWorkedWith'].str.contains('Python', na=False).sum()

# 3. Рахуємо відсоток від ЗАГАЛЬНОЇ кількості
python_percentage = (python_users_count / total_respondents) * 100

# 4. Виводимо результат
print(f"Кількість розробників на Python: {python_users_count}")
print(f"Відсоток від загальної кількості респондентів: {python_percentage:.1f}%")
```

```python
# Завдання 6. Аналіз шляхів навчання програмуванню (Online Courses)

# Використовуємо case=False, щоб не зважати на велику/маленьку літеру
# та na=False, щоб проігнорувати порожні клітинки
online_courses_count = df['LearnCode'].str.contains('Online Courses', case=False, na=False).sum()

# Виводимо результат
print(f"Кількість респондентів, які обрали онлайн-курси: {online_courses_count}")
```

```python
# Завдання 7. Географічний аналіз компенсації Python-розробників

# 1. Відфільтровуємо лише тих, хто програмує на Python
python_devs = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)].copy()

# 2. Групуємо за країною та розраховуємо середнє і медіану для ConvertedCompYearly
# Використовуємо метод .agg() для одночасного розрахунку обох показників
country_compensation = python_devs.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])

# 3. Перейменовуємо колонки для зручності та сортуємо за медіаною (від найбільшої)
country_compensation = country_compensation.rename(columns={
    'mean': 'Середня компенсація',
    'median': 'Медіанна компенсація'
}).sort_values(by='Медіанна компенсація', ascending=False)

# 4. Виводимо перші 15 країн (найбільш високооплачуваних)
print("Аналіз компенсації Python-розробників по країнах:")
country_compensation.head(15)
```

```python
# Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

# 1. Створюємо вибірку, де видаляємо рядки без зарплати
# Це гарантує, що ми працюємо лише з реальними цифрами
top_paid_devs = df.dropna(subset=['ConvertedCompYearly'])

# 2. Сортуємо за компенсацією від більшої до меншої (ascending=False)
# Та вибираємо перші 5 записів за допомогою head(5)
top_5_compensation = top_paid_devs.sort_values(by='ConvertedCompYearly', ascending=False).head(5)

# 3. Вибираємо лише необхідні колонки для відображення: освіта та сума
result = top_5_compensation[['ConvertedCompYearly', 'EdLevel']]

# 4. Виводимо результат
print("Рівні освіти 5 респондентів з найбільшою компенсацією:")
print(result)
```

```python
# Завдання 9. Аналіз популярності Python по вікових категоріях (Бонус)

# 1. Створюємо нову тимчасову колонку-прапорець (True/False)
# Вона показує, чи працює респондент з Python
df['Uses_Python'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

# 2. Групуємо за віком ('Age') та рахуємо середнє для колонки 'Uses_Python'
# Оскільки True = 1, а False = 0, середнє значення якраз і буде часткою (відсотком)
age_python_analysis = df.groupby('Age')['Uses_Python'].mean() * 100

# 3. Перетворюємо результат у гарний DataFrame
age_python_df = age_python_analysis.reset_index()
age_python_df.columns = ['Вікова категорія', 'Відсоток Python-розробників']

# 4. Сортуємо за відсотком, щоб побачити лідерів
age_python_df = age_python_df.sort_values(by='Відсоток Python-розробників', ascending=False)

# 5. Виводимо результат
print("Популярність Python за віковими групами:")
age_python_df
```

```python
# Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

# 1. Визначаємо поріг 75-го перцентиля для компенсації
# (quantile(0.75) знайде суму, вище якої заробляють лише 25% респондентів)
salary_threshold = df['ConvertedCompYearly'].quantile(0.75)

# 2. Фільтруємо за двома умовами:
# - Зарплата >= порогової
# - Формат роботи == 'Remote'
high_earning_remote = df[
    (df['ConvertedCompYearly'] >= salary_threshold) & 
    (df['RemoteWork'] == 'Remote')
].copy()

# 3. Підраховуємо частоту індустрій у цій групі
# Використовуємо value_counts(), що автоматично сортує від найпопулярніших
top_industries = high_earning_remote['Industry'].value_counts()

# 4. Виводимо результати
print(f"Поріг 75-го перцентиля зарплати: {salary_threshold:.0f}")
print(f"Кількість людей у цій групі: {len(high_earning_remote)}")
print("\nНайпопулярніші індустрії серед топ-оплачуваних remote-фахівців:")
print(top_industries.head(10))
```
